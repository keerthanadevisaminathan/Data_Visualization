```python
import pandas as pd
import matplotlib.pyplot as plt
Gender = data['Gender']
Nationality = data['Nationality']
indian_counts = data[data['Nationality'] == 'Indian']['Gender'].value_counts()
print("Number of males and females who are Indians:")
print(indian_counts)
indian_counts.plot(kind='bar', color='b')
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Number of Males and Females Who Are Indians")
plt.show()
```
